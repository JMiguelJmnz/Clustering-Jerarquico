```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

os.chdir("E:/EBAC/Material/M26")
data = pd.read_excel('Amazon.xlsx')
data.rename(columns={'Unnamed: 0': 'Cliente'}, inplace=True)

data.head()
```

```python
# separamos clientes de datos para normalizar datos
data_clients = data['Cliente']
only_data = data.drop(columns=['Cliente'])
only_data
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(only_data)
data_scaled
```

```python
name_cols = ['Velocidad Entrega', 'Precio', 'Durabilidad', 'Imagen Producto', 'Valor Educativo', 'Servicio Retorno',  'Tamano Paquete', 'Calidad Producto', 'Numero Estrellas'] 
data_scaled = pd.DataFrame(data_scaled, columns = name_cols)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("dendograma")
dend = shc.dendrogram(shc.linkage(data_scaled, method="ward"))
```

```python
print(dend['color_list'])
```

```python
colores_unicos = set(dend['color_list'])
colores_unicos
```

```python
num_clusters_optimo = len(colores_unicos) - 1
num_clusters_optimo
```

```python
plt.figure(figsize=(10,7))
plt.title("Dendograma")
dend = shc.dendrogram(shc.linkage(data_scaled, method = 'ward'))
plt.axhline(y=3, color = 'r', linestyle= '--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
grupos = cluster.fit_predict(data_scaled)
grupos
```

```python
# SeparaciÃ³n de grupos mediante PCA
campos = data_scaled.values
print(campos)
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca.fit(campos)
campos = pca.transform(campos)
campos
```

```python
plt.figure(figsize=(10,7))
plt.scatter(campos[:,0], campos[:,1], c = cluster.labels_)
```

```python
dataframe = pd.DataFrame(grupos, columns=['grupo'])
dataframe
```

```python
dataframe2 = pd.concat([data, dataframe], axis = 1, join = 'inner')
dataframe2
```
